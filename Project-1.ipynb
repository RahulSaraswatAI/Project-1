```python
cols=['dteday','season','yr','mnth','hr','weekday','weathersit','temp','atemp','hum','windspeed','casual','registered','cnt']
Final_Dataset[cols]=Final_Dataset[cols].apply(pd.to_numeric,args=('coerce',))
```

```python
q1=Final_Dataset.quantile(.25)
q3=Final_Dataset.quantile(.75)
```

```python
def find_outliers(df,cols):
    q1=df[cols].quantile(.25)
    q3=df[cols].quantile(.75)
    iqr=q3-q1
    min_r=q1-1.5*iqr
    max_r=q3+1.5*iqr
    outlier_indices=df.index[(df[cols]<min_r)|(df[cols]>max_r)]
    return outlier_indices
```

```python
find_outliers(Final_Dataset,'temp')
```

```python
find_outliers(Final_Dataset,'atemp')
```

```python
find_outliers(Final_Dataset,'hum')
```

```python
find_outliers(Final_Dataset,'windspeed')
```
